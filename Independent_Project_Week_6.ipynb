```
# We need to extract year and month from fifa_ranking

import datetime

fifa_ranking['Year'] = pd.DatetimeIndex(fifa_ranking['rank_date']).year
fifa_ranking['Month'] = pd.DatetimeIndex(fifa_ranking['rank_date']).month


fifa_ranking.head()
```

## Merging the Two Datasets

```
# Merging the two datasets using home_team

fifa_ranking_home = pd.merge(fifa_ranking, results, how='inner', left_on=['country_full','Year','Month'], right_on = ['home_team','Year', 'Month'] )
```

# Feature Engineering

```

fifa_merged.head()
```

```
# We need to determine whether the game is a win, loss or draw

fifa_merged['Goal_Difference'] = fifa_merged['Home_Score'] - fifa_merged['Away_Score']

def win_loss_draw(x):
  if x > 0:
    return 'Win'
  elif x < 0:
    return 'Loss'
  else:
    return 'Draw'

fifa_merged['Win_Loss_Draw'] = fifa_merged.Goal_Difference.apply(lambda x: win_loss_draw(x))
```

```
fifa_merged.head()
```

```
# Lable Encode to turn the Win_Loss_Draw and tournament column into numerical variables

from sklearn.preprocessing import LabelEncoder

en = LabelEncoder()

fifa_merged['Win_Loss_Draw'] = en.fit_transform(fifa_merged['Win_Loss_Draw'])
fifa_merged['Tournament'] = en.fit_transform(fifa_merged['Tournament'])

fifa_merged.head()
```

# Polynomial Model

## Approach 1: Model 1

```
# We need to separate our indepedent variables from the dependent variables

X = fifa_merged[['Home_Rank', 'Away_Rank', 'Tournament']]
y = fifa_merged['Home_Score']
```

## Checking for MultiColinearity

```
# Checking for correlation between the independent variables

corr_fifa = X.corr()

corr_fifa
```

```
# Checking the VIF for the independent variables

vif_fifa = pd.DataFrame(np.linalg.inv(corr_fifa.values), index = corr_fifa.index, columns = corr_fifa.columns)

vif_fifa
```

> Since all the independent variables have VIFs less than 5, we determine that they are suitable for our model.

## Model 1

```
X = fifa_merged[['Home_Rank', 'Away_Rank', 'Tournament']].values

y = fifa_merged['Home_Score'].values
```

```
# Spliting our data into traning and testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```
# Fitting the Polynomial regression

from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression

poly_fit = PolynomialFeatures(degree = 4)

X_train = poly_fit.fit_transform(X_train)
X_test = poly_fit.fit_transform(X_test)

poly_reg = LinearRegression()  # Calling the Linear Regression class into an object

poly_reg.fit(X_train, y_train) # Training model
```

```
# We need to check out intercept and coefficients

print(poly_reg.intercept_)

print("\n")

print(poly_reg.coef_)
```

```
# Predicting Home_Scores

home_scores_predicted = poly_reg.predict(X_test)

home_scores_predicted.round(0)
```

## Evaluating Polynomial Model 1

```
# Evaluating model using RMSE

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, home_scores_predicted))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, home_scores_predicted))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, home_scores_predicted)))
```

```
# Computing the heteskedasticity 
# Plotting the residual plots

resid = home_scores_predicted - y_test

plt.scatter(home_scores_predicted, resid, color='magenta')
plt.title('Residual Plot')
plt.ylabel('Residuals')
plt.xlabel('Fitted Values')
plt.axhline(y = resid.mean(), color='black')
plt.show()
```

```
# Performing a cross validation using 5 folds

from sklearn.model_selection import KFold

X = fifa_merged[['Home_Rank', 'Away_Rank', 'Tournament']].values

y = fifa_merged['Home_Score'].values

folds = KFold(n_splits=5) # Determing our folds

RMSES = [] 
count = 1
for train_index, test_index in folds.split(X):
  print('\nTraining model ' + str(count)) 

  
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]

 

  poly_fit = PolynomialFeatures(include_bias = False, degree = 4)

  X_train = poly_fit.fit_transform(X_train)
  X_test = poly_fit.fit_transform(X_test)

  # Training the model
  poly_reg = LinearRegression()
  poly_reg.fit(X_train, y_train)

  home_scores_predicted = poly_reg.predict(X_test)

  rmse_value =  np.sqrt(metrics.mean_squared_error(y_test, home_scores_predicted))
  RMSES.append(rmse_value)
    
  print('Model ' + str(count) + ' Root Mean Squared Error:',rmse_value)
  count = count + 1
```

```
np.mean(RMSES)
```

>> We should pick Model 3 which has the lowest RMSE score. A low RMSE score means that the model has fewe errors.

# Logistic Model

```
# Previewing the dataset we are using for our model 

fifa_merged.head()
```

```
fifa_merged.columns
```

```
# Selecting our independent and dependent variables

X = fifa_merged[['Home_Score', 'Away_Score', 'Home_Rank',
                 'Away_Rank', 'Tournament',]]

y = fifa_merged['Win_Loss_Draw']
```

```
# Spliting the data set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=25)

# Fitting the model 

from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression()

LogReg.fit(X_train, y_train)
```

```
y_pred = LogReg.predict(X_test)

y_pred
```

## Evaluating the Model

```
# Evaluating the logistic model using confusion matrix

from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, y_pred)

confusion_matrix
```

```

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```
# Calculating RMSE

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
# Using a Scatter plot to check for heteroskedasticity

residuals = y_pred - y_test


plt.scatter(y_pred, residuals, color='magenta')
plt.title('Residual Plot')
plt.ylabel('Residual')
plt.xlabel('Fitted Values')
plt.axhline(y = residuals.mean(), color='black')
plt.show()
```

> Model has no heteroskedasticity since there are no errors.
